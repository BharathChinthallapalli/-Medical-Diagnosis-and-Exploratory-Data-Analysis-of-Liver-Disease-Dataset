#Processing liver patient data for further diagnosis purposes

```python
# Calculate average age
avg_age = data['Age of the patient'].mean()

# Calculate average Total Bilirubin level
avg_total_bilirubin = data['Total Bilirubin'].mean()

# Calculate percentage of patients with abnormal liver function
total_patients = len(data)
abnormal_liver_function = len(data[data['Result'] == 1])
percentage_abnormal_liver_function = (abnormal_liver_function / total_patients) * 100

# Print summary statistics
print('Summary Statistics:')
print('---------------------')
print(f'Average Age: {avg_age:.2f} years')
print(f'Average Total Bilirubin level: {avg_total_bilirubin:.2f} mg/dL')
print(f'Percentage of patients with abnormal liver function: {percentage_abnormal_liver_function:.2f}%')
```

# Prompt user for input

```python

print('\nPlease enter the following information for diagnosis:')
age_input = float(input('Age of the patient (in years): '))
gender_input = input('Gender of the patient (Male/Female): ')
total_bilirubin_input = float(input('Total Bilirubin level (mg/dL): '))
direct_bilirubin_input = float(input('Direct Bilirubin level (mg/dL): '))
alkaline_phosphatase_input = float(input('Alkaline Phosphatase level (U/L): '))
sgpt_input = float(input('Sgpt Alamine Aminotransferase level (U/L): '))
sgot_input = float(input('Sgot Aspartate Aminotransferase level (U/L): '))
total_proteins_input = float(input('Total Proteins level (g/dL): '))
albumin_input = float(input('Albumin level (g/dL): '))
a_g_ratio_input = float(input('A/G Ratio Albumin and Globulin Ratio: '))
```

#Diagnosis calculation using medical knowledge

```python
# Initialize diagnosis
diagnosis = ''

# Age
if age_input > avg_age:
    diagnosis += 'Elderly patient. '

# Gender
if gender_input.lower() == 'male':
    diagnosis += 'Male gender. '
elif gender_input.lower() == 'female':
    diagnosis += 'Female gender. '

# Total Bilirubin
if total_bilirubin_input > avg_total_bilirubin:
    diagnosis += 'Elevated Total Bilirubin level. '

# Direct Bilirubin
if direct_bilirubin_input > 0.3:
    diagnosis += 'Elevated Direct Bilirubin level. '

# Alkaline Phosphatase
if alkaline_phosphatase_input > 180.0:
    diagnosis += 'Elevated Alkaline Phosphatase level. '

# Sgpt Alamine Aminotransferase
if sgpt_input > 40.0:
    diagnosis += 'Elevated Sgpt Alamine Aminotransferase level. '

# Sgot Aspartate Aminotransferase
if sgot_input > 35.0:
    diagnosis += 'Elevated Sgot Aspartate Aminotransferase level. '

# Total Proteins
if total_proteins_input < 6.0:
    diagnosis += 'Low Total Proteins level. '

# Albumin
if albumin_input < 3.5:
    diagnosis += 'Low Albumin level. '

# A/G Ratio Albumin and Globulin Ratio
if a_g_ratio_input < 1.1:
    diagnosis += 'Low A/G Ratio Albumin and Globulin Ratio. '

# Overall Diagnosis
if diagnosis == '':
    diagnosis = 'No abnormalities detected. Patient appears to be in good health.'
else:
    diagnosis = 'Based on the provided information, the following potential health concerns were identified: ' + diagnosis

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import networkx as nx
from PIL import Image

# Load the dataset
data = pd.read_csv('LPD.csv', encoding='latin1')

# Data visualization: Age distribution by gender
sns.histplot(x='Age of the patient', hue='Gender of the patient', data=data, bins=20, kde=True)
plt.title('Age Distribution by Gender')
plt.ylabel('Frequency')
plt.xlabel('Age')
plt.show()

# Interactive Box Data visualization: Total bilirubin distribution by gender
fig = px.box(data, x='Gender of the patient', y='Total Bilirubin', color='Gender of the patient', 
             title='Total Bilirubin Distribution by Gender')
fig.update_layout(xaxis_title='Gender', yaxis_title='Total Bilirubin (mg/dL)')
fig.show()

# Interactive Histogram Data visualization: Age distribution by gender
fig = px.histogram(data, x='Age of the patient', color='Gender of the patient', nbins=20, 
                   marginal='violin', title='Age Distribution by Gender')
fig.update_layout(xaxis_title='Age', yaxis_title='Count')
fig.show()

# Create a network graph to visualize the relationships between parameters
G = nx.Graph()

# Add nodes
G.add_node('Age')
G.add_node('Gender')
G.add_node('Total Bilirubin')
G.add_node('Direct Bilirubin')
G.add_node('Alkaline Phosphatase')
G.add_node('Sgpt Alamine Aminotransferase')
G.add_node('Sgot Aspartate Aminotransferase')
G.add_node('Total Proteins')
G.add_node('Albumin')
G.add_node('A/G Ratio Albumin and Globulin Ratio')

# Add edges
G.add_edge('Age', 'Gender')
G.add_edge('Age', 'Total Bilirubin')
G.add_edge('Age', 'Direct Bilirubin')
G.add_edge('Gender', 'Total Bilirubin')
G.add_edge('Total Bilirubin', 'Direct Bilirubin')
G.add_edge('Total Bilirubin', 'Alkaline Phosphatase')
G.add_edge('Total Bilirubin', 'Sgpt Alamine Aminotransferase')
G.add_edge('Direct Bilirubin', 'Alkaline Phosphatase')
G.add_edge('Alkaline Phosphatase', 'Sgpt Alamine Aminotransferase')
G.add_edge('Alkaline Phosphatase', 'Sgot Aspartate Aminotransferase')
G.add_edge('Sgpt Alamine Aminotransferase', 'Sgot Aspartate Aminotransferase')
G.add_edge('Sgot Aspartate Aminotransferase', 'Total Proteins')
G.add_edge('Total Proteins', 'Albumin')
G.add_edge('Total Proteins', 'A/G Ratio Albumin and Globulin Ratio')
G.add_edge('Albumin', 'A/G Ratio Albumin and Globulin Ratio')

# Draw the network graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='skyblue', font_size=10, edge_color='gray')
plt.title('Parameter Relationships')
plt.show()

# Save the network graph as an image
```
